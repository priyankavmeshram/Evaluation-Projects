# checking for the outliers

```python
plt.figure(figsize= (20,15), facecolor='white')
plotnumber = 1

for column in numerical_col:
    if plotnumber<15:
        ax = plt.subplot(3,5, plotnumber)
        sns.boxplot(df[column], palette = 'Set2_r')
        plt.xlabel(column, fontsize=20)
        plt.yticks(rotation = 0, fontsize = 20)
    plotnumber+=1
plt.tight_layout()
plt.show()
```

As we can see the outliers are present  in the column age, ploicy annual premium and umbrella limit. lets remove these outliers usin zscore method or Iqr technique.

```python
from scipy.stats import zscore
out_features = df[['age','policy_annual_premium','umbrella_limit']]
z= np.abs(zscore(out_features))
z
```

```python
#threshold=3
np.where(z>3)
```

```python
#so now removing the data above zscore and creating a new df
df1= df[(z<3).all(axis=1)]
df1.shape
```

```python
print("Data loss percentage- ",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
#now using IQR method

#1st quantile
Q1 = out_features.quantile(0.25)

#3rd quantile
Q3 = out_features.quantile(0.75)

#IQR
IQR = Q3 - Q1

df2= df[~((df <(Q1 - 1.5 * IQR))|(df> (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df2.shape
```

```python
print("Data loss percentage- ",((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

As we see that the zscore method will give the less amount of loss percentage data so we used df1 as our new dataframe

```python
df=df1
```

# Modelling

Finding the best random State

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu = 0
maxRs=0
for i in range(1,200):
    x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.30, random_state = i)
    RFR = RandomForestClassifier()
    RFR.fit(x_train, y_train)
    pred = RFR.predict(x_test)
    acc = accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRs= i
print("Best accuracy is", maxAccu, "at random_state", maxRs)
```

the best accuracy score is 92.15% at random state 14

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30, random_state= maxRs)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```
